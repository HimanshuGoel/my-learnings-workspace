```python
import numpy as np
import time

A=np.random.rand(2000,2000)
B=np.random.rand(2000,2000)

start=time.time()

C=np.matmul(A,B)
end=time.time()
print("CPU time: ", end-start)
```

```python
end=time.time()
print("CPU time: ", end-start)
```

```python
import torch
import time


device="cuda"
A=torch.rand(2000,2000).to(device)
B=torch.rand(2000,2000).to(device)

start=time.time()
C=torch.matmul(A,B)
end=time.time()
print("GPU Time: ", end-start)

```
